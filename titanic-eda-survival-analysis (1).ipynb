```python
df = pd.read_csv('/kaggle/input/competitions/titanic/train.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df["Survived"].value_counts()
```

```python
df.groupby("Sex")["Survived"].mean()
```

```python
df.groupby("Pclass")["Survived"].mean()
```

```python
df.groupby(["Pclass","Sex"])["Survived"].mean()
```

```python
df['Age'].median()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df.isnull().sum()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
df[['SibSp', 'Parch', 'FamilySize']].head()
```

```python
df.groupby('FamilySize')['Survived'].mean()
```

```python
fare_array = df['Fare'].to_numpy()
print("Mean:", np.mean(fare_array))
print("Std Dev:", np.std(fare_array))
print("Max:", np.max(fare_array))
print("Min:", np.min(fare_array))
```

```python
print("Median Fare:", np.median(fare_array))
```

```python
pd.pivot_table(df, values='Survived', index='Pclass', columns='Sex', aggfunc='mean')
```

```python
df[['Survived', 'Age', 'Fare', 'Pclass', 'FamilySize']].corr()
```

```python
df['AgeGroup'] = np.where(df['Age'] < 18, 'Child', 'Adult')
df.groupby('AgeGroup')['Survived'].mean()
```
